**#MACHINE LEARNING**

```python
#Dùng file mtcars.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model #thư viện dùng cho machine learning - mô hình tuyến tính

url = '/content/drive/MyDrive/Colab Notebooks/AI_Projects/Data/mtcars.csv'
data = pd.read_csv(url)
data.hist()  #vẽ phân bố dữ liệu Histogram
plt.show
     
```

**VD1: Hồi quy 1  biến**

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
#giá trị tuyệt đối càng lớn thì mối tương quan càng lớn
```

```python
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black')
```

```python
#Hồi quy hàm bậc 1
model = linear_model.LinearRegression()  #hàm hồi quy tuyến tính

X=pd.DataFrame(data['wt'])
y=data['mpg']

model.fit(X, y) #chạy để nó học
print(model.intercept_)
print(model.coef_)
#PT hồi quy : y = coef*X + intercept

print(model.score(X, y))

data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=0.7)
y_pred = model.predict(X)
plt.plot(data['wt'], y_pred, color='blue')
```

```python
#Hồi quy bậc 2
model = linear_model.LinearRegression()
X=pd.DataFrame([data['wt'], data['wt']**2]).T #T : chuyển vị
y=data['mpg']
#PT hồi quy : y = a1*x + a2*x^2 + b
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['wt'].min(), data['wt'].max(), 0.1)
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```

```python
#Hồi quy bậc 3
model = linear_model.LinearRegression()
X=pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T #T : chuyển vị
y=data['mpg']
#PT hồi quy : y = a1*x + a2*x^2 + a3*x^3 + b
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['wt'].min(), data['wt'].max(), 0.01)
predictor = pd.DataFrame([range, range**2, range**3]).T
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=0.7)
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```

```python
#Hồi quy bậc 5
model = linear_model.LinearRegression()
X=pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, data['wt']**4, data['wt']**5]).T #T : chuyển vị
y=data['mpg']
#PT hồi quy : y = a1*x + a2*x^2 + a3*x^3 + b
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['wt'].min(), data['wt'].max(), 0.1)
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
     
```

```python
#Hồi quy bậc 10
model = linear_model.LinearRegression()
X=pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, data['wt']**4, data['wt']**5,
                data['wt']**6, data['wt']**7, data['wt']**8, data['wt']**9, data['wt']**10]).T #T : chuyển vị
y=data['mpg']
#PT hồi quy : y = a1*x + a2*x^2 + a3*x^3 + b
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['wt'].min(), data['wt'].max(), 0.1)
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6, range**7, range**8, range**9, range**10]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```

> Vậy ta thấy mô hình bậc 4 là mô hình có tỉ lệ chính xác nhất
